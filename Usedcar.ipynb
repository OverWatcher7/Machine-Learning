```python
import numpy as np
import pandas as pd 
import matplotlib
import matplotlib.pyplot as plt 
import sklearn
```

```python
data=pd.read_csv("C:\\Users\\PC\\Desktop\\Usedcar.csv",delimiter=',')
print(data)
```

```python
x=np.c_[data['Mileage']]
y=np.c_[data['Price']]
```

```python
data.plot(kind="scatter",x="Mileage",y="Price")
plt.show()
```

```python
from sklearn import linear_model
lr_model=linear_model.LinearRegression()
lr_model.fit(x,y)
print("斜率：%s，截距：%s"%(lr_model.coef_[0][0],lr_model.intercept_[0]))
print("估计模型为：y=%sx+%s"%(lr_model.coef_[0][0],lr_model.intercept_[0]))
```

```python
data.plot(kind="scatter",x="Mileage",y="Price")
plt.plot(x,lr_model.predict(x.reshape(-1,1)),color='red',linewidth=4)
plt.show()
```
